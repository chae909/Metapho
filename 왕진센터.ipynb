## 1. 병원정보

```python
import pandas as pd

path = '/content/drive/MyDrive/Colab Notebooks/메타포3조/메타포DA/데이터/전국 병의원 및 약국 현황/1.병원정보서비스 2024.9.xlsx'

# 엑셀 파일 읽기
hosp_info = pd.read_excel(path, engine='openpyxl')

# hosp_info로 저장
hosp_info.to_csv('hosp_info.csv')

# 데이터프레임 출력
print(hosp_info.head())  # 처음 5행
```

```python
# '왕진'이 포함된 병원 필터링
wangjin = hosp_info[hosp_info['요양기관명'].str.contains('왕진', na=False)]

# 결과 출력
print(f"왕진 서비스 제공 병원 수: {len(wangjin)}")
print("\n왕진 서비스 제공 병원 목록:")
print(wangjin[['요양기관명', '종별코드명', '시군구코드명', '좌표(X)', '좌표(Y)']])  # 필요한 컬럼만 선택하여 출력
```

### 파주

```python
# 파주 데이터 pj_hosp에 분류
index1=hosp_info[hosp_info['시군구코드명'] != '파주시'].index
pj_hosp=hosp_info.drop(index1)
print(pj_hosp)

# 분류한 자료의 인덱스 정리
pj_hosp.reset_index(drop=True, inplace=True)
print(pj_hosp)
```

### 광명
